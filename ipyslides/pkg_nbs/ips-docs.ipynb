````python

skipper = slides.link('Skip to dynamic content', 'Back to link info', icon='arrow', back_icon='arrowl')

def _snapshots(s):
    grp = slides.group(snapshots=True)
    grp.append(['# Main App', slides.doc(isd.Slides)])

    with grp.capture():
        slides.write('# Jump between slides')
        slides.doc(slides.link, 'Slides').display()
        with slides.code.context(returns=True) as c:
            skipper.origin.display()
        c.display()

    grp.append([
        '## Adding Slides section`Adding Slides and Content`',
        'Besides function below, you can add slides with `%%slide number [-m]` magic as well.\n{.note .info}',
        slides.doc(slides.slide, 'Slides'),
        slides.doc(slides.sync_with_file, 'Slides'),
    ])

    with grp.capture():
        slides.write('''
            ## Important Methods on Slide
            ::: note-warning
                - Use slide handle or `Slides[number,]` to apply these methods because index can change on new builds.
                - Use `Slides[start:stop:step]` to apply operations on many slides at once such as code`Slides[2:5].vars.update(...)`.
        ''')
        slides.doc(slides[0], members='vars update_display get_source show'.split(), itself=False).display()

    slides.pause()
    slides.write(grp)

with slides.slide(2):
    slides.pending(_snapshots) # lazy build
````

## 2) Content Authoring and Syntax
This section covers xmd syntax, content writing APIs, dual-usage methods, notes, and source-code display helpers.

````python
with slides.slide(3):
    slides.xmd.syntax.display()

def _shots2(s):
    grp = slides.group(snapshots=True)
    with grp.capture():
        slides.xmd('''
            ## Adding Content
            Besides functions below, you can add content to slides with `%%xmd`, `%xmd` as well.
            {.note .info}
        ''')
        slides.write(
            slides.doc(slides.write, 'Slides'),
            [
                slides.doc(slides.xmd, 'Slides'),
                slides.doc(slides.as_html, 'Slides'),
                slides.doc(slides.as_html_widget, 'Slides'),
                slides.doc(slides.html, 'Slides')
            ]
        )

    grp.append(['''
        ## Dual Usage Functions
        Some functions such as `Slides.css`, `Slides.yoffset`, and `Slides.transition` can be used
        both inside and outside slide context through their `applyto` parameter.
        If `applyto` is `None` (or omitted), they require an active slide context and apply to the current slide.
        If `applyto` is provided (for example, a slide index, list of indices, or `'all'`), they can be used outside
        a slide context and apply to the selected slides. They can also be used from markdown with syntax like `function[applyto]\`content\``.
        ''',
        slides.doc(slides.transition, 'Slides'),
        slides.doc(slides.yoffset, 'Slides'),
        slides.doc(slides.css, 'Slides'),
        slides.css_syntax,
    ])

    grp.append(['''
        ## Adding Speaker Notes
        styled["note success"]`You can use alert`notes\`notes content\`` in markdown.`
        ::: note-error error | This is experimental feature, and may not work as expected.
        ''',
        slides.doc(slides.notes, 'Slides.notes', members=True, itself=False)
    ])

    grp.append([
        '## Displaying Source Code',
        'In markdown, the block `md-[before,after,var_name]` parses and displays source as well.',
        slides.doc(slides.code, 'Slides', members=True, itself=True),
    ])

    slides.pause() # pause before write command makes incremental updates
    slides.write(grp)

with slides.slide(4):
    slides.pending(_shots2)
````

## 4) Dynamic Content, Styling, and Export
These cells preserve the interactive parts of the old docs function: dashlab interact, on-load hooks, content styling, animations, highlighting, and export workflow.

````python
%%slide 9
@slides.pending
def _dy_content(s):
    skipper.target.display()
    slides.write('## Dynamic Content')

    with slides.capture_content() as cap, slides.code.context():
        import time

        @slides.dl.interact(
            post_init=lambda self: self.set_css(
                dict({'.out-main': dict(height='2em')}, background='var(--bg2-color)')
            ),
            date=False,
            btn=slides.dl.button(),
        )
        def update_time(date, btn):
            local_time = time.localtime()
            objs = ['Time: {3}:{4}:{5}'.format(*local_time)]
            if date:
                objs.append('Date: {0}/{1}/{2}'.format(*local_time))
            slides.stack(objs).display()

    with slides.code.context(returns=True) as c:
        import datetime

        @slides.on_load
        def push_toast(slide):
            t = datetime.datetime.now()
            now = t.strftime('%H:%M:%S')
            slides.notify(f'Notification at {now} from slide {slide.index} and frame {slide.indexf}', timeout=5)
            return lambda ss: slides.notify('x')

    slides.write(
        [slides.doc(slides.dl.interact, 'ipyslides.dashlab')],
        [*cap.outputs, c, slides.doc(slides.on_load, 'Slides')],
    )
````

## 5) Advanced Functionality
The final section covers incremental frames, serializer/extensions, focus behavior, icons, and auto-numbering notes.

````python
slides.slide(14, 'section`Advanced Functionality` toc`### Contents`')

with slides.slide(15):
    @slides.pending
    def _(s):
        slides.write('## Adding content on frames incrementally')
        slides.frozen(widget := (code := s.get_source()).as_widget()).display()

        def highlight_code(slide):
            widget.value = code.focus(range(slide.indexf + 1)).value

        slides.on_load(highlight_code)

        for ws, cols in slides.pause.iter(
            zip([None, (2, 3), None], [(0, 1), (2, 3), (4, 5, 6, 7)]),
            isolate=True,
        ):
            cols = [
                slides.html('h1', f"{c}", style='background:var(--bg3-color);margin-block:0.05em !important;')
                for c in cols
            ]
            slides.write(*cols, widths=ws, css_class='anim-group anim-wipe-right')

with slides.slide(16):
    @slides.pending
    def _(s):
        slides.write('## Adding User defined Objects/Markdown Extensions')
        slides.write(
            slides.hold(
                display,
                slides.html('h3', 'I will be on main slides', css_class='warning'),
                metadata={'text/html': '<h3 class="warning">I will be on exported slides</h3>'},
            ),
            s.get_source(),
            widths=[1, 3],
        )
        slides.write(r'If you need to serialize your own or third party objects not serialized by this module, you can use `\@Slides.serializer.register` to serialize them to html.\n{.note .info}')
        slides.doc(slides.serializer, 'Slides.serializer', members=True, itself=False).display()
        slides.write('**You can also extend markdown syntax** using `markdown extensions`, ([See here](https://python-markdown.github.io/extensions/) and others to install, then use as below):')
        slides.doc(slides.xmd.extensions, 'Slides.xmd.extensions', members=True, itself=False).display()

with slides.slide(17):
    slides.write(r'''
    ## Focus on what matters
    - Most of supported elements can be focused by default like images, matplotlib, bokeh, PIL image, altair plotly, dataframe, etc.
    - You can also enable focus on an object/widget by wrapping it inside `Slide.focus` function conveniently.
    - You can also enable focus by manually adding `focus-self`, `focus-child` classes to an element.

    ::: focus-self block-red
        ### Focus on Me 😎
        Double click to focus on this block. Click at top right button or double click to exit.
    ''')

with slides.slide(18):
    @slides.pending
    def _(s):
        with slides.capture_content() as c:
            with slides.code.context():
                import ipywidgets as ipw
                btn = ipw.Button(description='Chevron-Down Icon', icon='chevrond')
                slides.write(btn)

        group = zip(slides.icon.available[::2], slides.icon.available[1::2])
        slides.write(
            ['''
                ## SVG Icons
                Icons that appear on buttons in slides (and their rotations) are available to use in your slides as well
                besides standard ipywidgets icons.
                ''', *c.outputs, 'line`200`**Source code of this slide**', slides.this.get_source()],
            slides.table(
                [(f'`{j}`', slides.icon(j, color='crimson').svg, f'`{k}`', slides.icon(k, color='crimson').svg) for j, k in group],
                headers=['Name', 'Icon', 'Name', 'Icon'],
            ),
            widths=[3, 2],
        )

with slides.slide(19):
    slides.write("""
        # Auto Slide Numbering
        Use alert`-1` as placeholder to update slide number automatically.

        - In Jupyter notebook, this will be updated to current slide number.
        - In python file, it stays same.
        - You may need to run cell twice if creating slides inside a for loop while using `-1`.

        ::: note-warning
            Some kernels may not support auto slide numbering inside notebook.
    """)
````
